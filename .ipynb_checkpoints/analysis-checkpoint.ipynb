### Unsupervised Learning: PCA and KMeans
---

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Load dataset
df = pd.read_csv("cleaned_data.csv")

# Create quartile-based target column
df['wae_mod_cat'] = pd.qcut(df['wae_mod'], 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

# Columns to exclude
exclude_cols = [
    'lake_nhdid', 'lake_namelagos', 'merge_key',
    'state', 'state_abbr', 'county_name',
    'lake_countyfips', 'fips', 'wae_mod', 'wae_mod_cat', 'year_diff', 'wae_mod_lifestage'
]

# Prepare features
X = df.drop(columns=exclude_cols)
X = pd.get_dummies(X, drop_first=True)

# Standardize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
df['PCA1'] = X_pca[:, 0]
df['PCA2'] = X_pca[:, 1]

# KMeans
kmeans = KMeans(n_clusters=4, random_state=42)
df['cluster'] = kmeans.fit_predict(X_pca)

# Plot side-by-side
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Plot 1: PCA colored by cluster
scatter1 = axs[0].scatter(df['PCA1'], df['PCA2'], c=df['cluster'], cmap='viridis', s=50)
axs[0].legend(*scatter1.legend_elements(), title="Cluster")
axs[0].set_title('PCA Colored by KMeans Cluster')
axs[0].set_xlabel('Principal Component 1')
axs[0].set_ylabel('Principal Component 2')
axs[0].grid(True)

# Plot 2: PCA colored by wae_mod_cat
scatter2 = axs[1].scatter(df['PCA1'], df['PCA2'], c=pd.Categorical(df['wae_mod_cat']).codes, cmap='viridis', s=50)
axs[1].legend(*scatter2.legend_elements(), title="Success Quartile")
axs[1].set_title('PCA Colored by Quartile')
axs[1].set_xlabel('Principal Component 1')
axs[1].set_ylabel('Principal Component 2')
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

### First Attempt: Decision Tree Regressor and Classifier

---

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.metrics import (
    confusion_matrix, ConfusionMatrixDisplay,
    classification_report, accuracy_score,
    precision_score, recall_score, f1_score,
    mean_squared_error, r2_score
)

# Load data
df = pd.read_csv("cleaned_data.csv")
df['wae_mod_cat'] = pd.qcut(df['wae_mod'], 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

X = df.drop(columns=exclude_cols)
X = pd.get_dummies(X, drop_first=True)

# Targets
y_class = df['wae_mod_cat']
y_reg = df['wae_mod']

# Train/test splits
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X, y_class, test_size=0.2, random_state=42)
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(X, y_reg, test_size=0.2, random_state=42)

# --- Classifier ---
clf = DecisionTreeClassifier(random_state=42)
param_grid_c = {'max_depth': [3, 5, 7, 10, None], 'min_samples_split': [2, 5, 10]}
grid_clf = GridSearchCV(clf, param_grid_c, cv=5)
grid_clf.fit(X_train_c, y_train_c)
y_pred_c = grid_clf.predict(X_test_c)

# Classification metrics
print("Decision Tree Classifier Performance")
print(f"Best Parameters: {grid_clf.best_params_}")
print("Accuracy:", accuracy_score(y_test_c, y_pred_c))
print("Precision (macro):", precision_score(y_test_c, y_pred_c, average='macro'))
print("Recall (macro):", recall_score(y_test_c, y_pred_c, average='macro'))
print("F1 Score (macro):", f1_score(y_test_c, y_pred_c, average='macro'))
print("\nClassification Report:\n", classification_report(y_test_c, y_pred_c))

# Confusion Matrix
fig, ax = plt.subplots(figsize=(6, 6))
ConfusionMatrixDisplay.from_predictions(y_test_c, y_pred_c, ax=ax, cmap='Blues', xticks_rotation=45)
plt.title("Confusion Matrix - Decision Tree Classifier")
plt.tight_layout()
plt.show()

# --- Regressor ---
reg = DecisionTreeRegressor(random_state=42)
param_grid_r = {'max_depth': [3, 5, 7, 10, None], 'min_samples_split': [2, 5, 10]}
grid_reg = GridSearchCV(reg, param_grid_r, cv=5)
grid_reg.fit(X_train_r, y_train_r)
y_pred_r = grid_reg.predict(X_test_r)

# Regression metrics
mse = mean_squared_error(y_test_r, y_pred_r)
r2 = r2_score(y_test_r, y_pred_r)
print("\nDecision Tree Regressor Performance")
print(f"Best Parameters: {grid_reg.best_params_}")
print(f"Mean Squared Error: {mse:.4f}")
print(f"R² Score: {r2:.4f}")
```

The decision tree classifier had a weak overall performance accurately predicting less than 30% for Q1, Q2, and Q3. The model performed better with Q4, achieving a precision of 47%. This suggests that Q4 has distinctive feature patterns. The decision tree regressor underperformed as the model explains only 12.16% of the variance of wae_mod. This indicated that the model is too simple to capture the complex underlying relationships between the features and wae_mod.

### Second Attempt: Random Forest Regressor and Classifier

---

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import (
    classification_report, accuracy_score, precision_score,
    recall_score, f1_score, ConfusionMatrixDisplay,
    mean_squared_error, r2_score
)

# Load data
df = pd.read_csv("cleaned_data.csv")
df['wae_mod_cat'] = pd.qcut(df['wae_mod'], 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

X = df.drop(columns=exclude_cols)
X = pd.get_dummies(X, drop_first=True)

y_class = df['wae_mod_cat']
y_reg = df['wae_mod']

# Train-test splits
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X, y_class, test_size=0.2, random_state=42)
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(X, y_reg, test_size=0.2, random_state=42)

# --- Random Forest Classifier ---
clf = RandomForestClassifier(random_state=42)
param_grid_c = {
    'n_estimators': [100],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5]
}
grid_clf = GridSearchCV(clf, param_grid_c, cv=5)
grid_clf.fit(X_train_c, y_train_c)
y_pred_c = grid_clf.predict(X_test_c)

# Classifier Metrics
print("Random Forest Classifier")
print("Best Parameters:", grid_clf.best_params_)
print("Accuracy:", accuracy_score(y_test_c, y_pred_c))
print("Precision (macro):", precision_score(y_test_c, y_pred_c, average='macro'))
print("Recall (macro):", recall_score(y_test_c, y_pred_c, average='macro'))
print("F1 Score (macro):", f1_score(y_test_c, y_pred_c, average='macro'))
print("\nClassification Report:\n", classification_report(y_test_c, y_pred_c))

# Confusion Matrix
fig, ax = plt.subplots(figsize=(6, 6))
ConfusionMatrixDisplay.from_predictions(y_test_c, y_pred_c, ax=ax, cmap='Blues', xticks_rotation=45)
plt.title("Confusion Matrix - Random Forest Classifier")
plt.tight_layout()
plt.show()

# --- Random Forest Regressor ---
reg = RandomForestRegressor(random_state=42)
param_grid_r = {
    'n_estimators': [100],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5]
}
grid_reg = GridSearchCV(reg, param_grid_r, cv=5)
grid_reg.fit(X_train_r, y_train_r)
y_pred_r = grid_reg.predict(X_test_r)

# Regressor Metrics
print("\nRandom Forest Regressor")
print("Best Parameters:", grid_reg.best_params_)
print("Mean Squared Error:", mean_squared_error(y_test_r, y_pred_r))
print("R² Score:", r2_score(y_test_r, y_pred_r))
```

The random forest classifier outperformed the decision tree regressor, increasing the precision across quantiles. Q4 again stood out as the easiest to predict. The random forest regressor also faired better than the decision tree regressor. The model explained 34.75% of the variance in wae_mod. While low, it is a signifcant improvement from the previous regressor model.

### Third Attempt: XGBoost Classifier and Regressor

```python
import pandas as pd
import matplotlib.pyplot as plt
from xgboost import XGBClassifier, XGBRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    classification_report, ConfusionMatrixDisplay,
    mean_squared_error, r2_score
)

# Load dataset
df = pd.read_csv("cleaned_data.csv")
df['wae_mod_cat'] = pd.qcut(df['wae_mod'], 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

X = df.drop(columns=exclude_cols)
X = pd.get_dummies(X, drop_first=True)

# Targets
y_class = df['wae_mod_cat']
y_reg = df['wae_mod']

# Encode classification target for XGBoost
y_class_encoded = y_class.astype('category').cat.codes
label_map = dict(enumerate(y_class.astype('category').cat.categories))

# Train/test splits
X_train_c, X_test_c, y_train_c, y_test_c = train_test_split(X, y_class_encoded, test_size=0.2, random_state=42)
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(X, y_reg, test_size=0.2, random_state=42)

# --- XGBoost Classifier ---
xgb_clf = XGBClassifier(eval_metric='mlogloss', random_state=42)
param_grid_xgb_c = {
    'max_depth': [3, 5],
    'learning_rate': [0.1],
    'n_estimators': [100]
}
grid_xgb_clf = GridSearchCV(xgb_clf, param_grid_xgb_c, cv=5)
grid_xgb_clf.fit(X_train_c, y_train_c)
y_pred_c = grid_xgb_clf.predict(X_test_c)

# Classifier metrics
print("XGBoost Classifier")
print("Best Parameters:", grid_xgb_clf.best_params_)
print("Accuracy:", accuracy_score(y_test_c, y_pred_c))
print("Precision (macro):", precision_score(y_test_c, y_pred_c, average='macro'))
print("Recall (macro):", recall_score(y_test_c, y_pred_c, average='macro'))
print("F1 Score (macro):", f1_score(y_test_c, y_pred_c, average='macro'))
print("\nClassification Report:\n", classification_report(y_test_c, y_pred_c, target_names=label_map.values()))

# Confusion Matrix
fig, ax = plt.subplots(figsize=(6, 6))
ConfusionMatrixDisplay.from_predictions(
    y_test_c, y_pred_c, display_labels=label_map.values(), ax=ax, cmap='Blues', xticks_rotation=45)
plt.title("Confusion Matrix - XGBoost Classifier")
plt.tight_layout()
plt.show()

# --- XGBoost Regressor ---
xgb_reg = XGBRegressor(random_state=42)
param_grid_xgb_r = {
    'max_depth': [3, 5],
    'learning_rate': [0.1],
    'n_estimators': [100]
}
grid_xgb_reg = GridSearchCV(xgb_reg, param_grid_xgb_r, cv=5)
grid_xgb_reg.fit(X_train_r, y_train_r)
y_pred_r = grid_xgb_reg.predict(X_test_r)

# Regressor metrics
print("\nXGBoost Regressor")
print("Best Parameters:", grid_xgb_reg.best_params_)
print("Mean Squared Error:", mean_squared_error(y_test_r, y_pred_r))
print("R² Score:", r2_score(y_test_r, y_pred_r))
```

In terms of classification, the XGBoost model performed similiarly to the random forest classifier. The prediction accuracy for Q4 increased again. The XGBoost regressor performed similarly to the random forest regressor as well.

### Final Attempt: XGBoost Classifier with Median Split in Success Category

---

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import (
    classification_report, accuracy_score, precision_score,
    recall_score, f1_score, ConfusionMatrixDisplay
)
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

# Load and prep data
df = pd.read_csv("cleaned_data.csv")
df['wae_mod_cat'] = pd.qcut(df['wae_mod'], 4, labels=['Q1', 'Q2', 'Q3', 'Q4'])

# New binary target: Top 50% (Q3 & Q4) vs Bottom 50% (Q1 & Q2)
df['is_top_half'] = df['wae_mod_cat'].isin(['Q3', 'Q4']).astype(int)

exclude_cols = [
    'lake_nhdid', 'lake_namelagos', 'merge_key',
    'state', 'state_abbr', 'county_name',
    'lake_countyfips', 'fips', 'wae_mod', 'wae_mod_cat', 'year_diff', 'wae_mod_lifestage', 'is_top_half']

X_bin = df.drop(columns=exclude_cols)
X_bin = pd.get_dummies(X_bin, drop_first=True)
y_bin = df['is_top_half']

# Split data
X_train_bin, X_test_bin, y_train_bin, y_test_bin = train_test_split(X_bin, y_bin, test_size=0.2, random_state=42)

# --- Random Forest Classifier ---
rf_bin = RandomForestClassifier(random_state=42)
param_grid_rf = {
    'n_estimators': [100],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5]
}
grid_rf = GridSearchCV(rf_bin, param_grid_rf, cv=5)
grid_rf.fit(X_train_bin, y_train_bin)
y_pred_rf = grid_rf.predict(X_test_bin)

print("\nRandom Forest - Top 50% Classifier")
print("Best Params:", grid_rf.best_params_)
print("Accuracy:", accuracy_score(y_test_bin, y_pred_rf))
print("Precision:", precision_score(y_test_bin, y_pred_rf))
print("Recall:", recall_score(y_test_bin, y_pred_rf))
print("F1 Score:", f1_score(y_test_bin, y_pred_rf))
print(classification_report(y_test_bin, y_pred_rf))

fig, ax = plt.subplots(figsize=(5, 5))
ConfusionMatrixDisplay.from_predictions(y_test_bin, y_pred_rf, ax=ax, cmap='Blues')
plt.title("Random Forest - Confusion Matrix (Top 50%)")
plt.tight_layout()
plt.show()

# --- XGBoost Classifier ---
xgb_bin = XGBClassifier(eval_metric='logloss', random_state=42)
param_grid_xgb = {
    'n_estimators': [100],
    'max_depth': [3, 5],
    'learning_rate': [0.1]
}
grid_xgb = GridSearchCV(xgb_bin, param_grid_xgb, cv=5)
grid_xgb.fit(X_train_bin, y_train_bin)
y_pred_xgb = grid_xgb.predict(X_test_bin)

print("\nXGBoost - Top 50% Classifier")
print("Best Params:", grid_xgb.best_params_)
print("Accuracy:", accuracy_score(y_test_bin, y_pred_xgb))
print("Precision:", precision_score(y_test_bin, y_pred_xgb))
print("Recall:", recall_score(y_test_bin, y_pred_xgb))
print("F1 Score:", f1_score(y_test_bin, y_pred_xgb))
print(classification_report(y_test_bin, y_pred_xgb))

fig, ax = plt.subplots(figsize=(5, 5))
ConfusionMatrixDisplay.from_predictions(y_test_bin, y_pred_xgb, ax=ax, cmap='Oranges')
plt.title("XGBoost - Confusion Matrix (Top 50%)")
plt.tight_layout()
plt.show()
```

With the redefined success class, the XGBoost classifier outperformed the random forest classifier.

```python
import numpy as np
import matplotlib.pyplot as plt

# Extract feature importances
xgb_importances = grid_xgb.best_estimator_.feature_importances_
top_xgb_idx = np.argsort(xgb_importances)[::-1][:15]

# Feature names from the training matrix
feature_names = X_bin.columns

# Plot side-by-side
plt.figure(figsize=(14, 6))

# XGBoost
plt.subplot(1, 2, 2)
plt.barh(range(15), xgb_importances[top_xgb_idx][::-1], color='orange')
plt.yticks(range(15), feature_names[top_xgb_idx][::-1])
plt.title("XGBoost - Top 15 Features (Top 50% Classifier)")
plt.xlabel("Importance")

plt.tight_layout()
plt.show()
```

```python
import shap
import matplotlib.pyplot as plt

# 1. Create the SHAP explainer 
explainer = shap.TreeExplainer(grid_xgb.best_estimator_)

# 2. Calculate SHAP values on the same input matrix used in training
shap_values = explainer.shap_values(X_bin)

# 3. Bar chart: Overall feature importance (mean absolute SHAP values)
shap.summary_plot(shap_values, X_bin, plot_type="bar")

# 4. Beeswarm plot: Feature impact across individual predictions
shap.summary_plot(shap_values, X_bin)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Get predicted probabilities (not just class labels)
y_probs = grid_xgb.predict_proba(X_test_bin)[:, 1]  # Probability of class 1

# ROC curve
fpr, tpr, thresholds = roc_curve(y_test_bin, y_probs)
auc_score = roc_auc_score(y_test_bin, y_probs)

# Plot
plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, label=f'XGBoost (AUC = {auc_score:.2f})')
plt.plot([0, 1], [0, 1], 'k--')  # Diagonal = random
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - XGBoost Top 50% Classifier')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
